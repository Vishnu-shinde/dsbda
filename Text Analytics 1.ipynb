```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
sw = stopwords.words('english')
def transform_text(s):
    tokens = nltk.word_tokenize(s)            # to split s into tokens
    
    new_string= []
    for w in tokens:
        if len(w)>2 and w not in sw:
            new_string.append(w)              # if token not in stopwords append to new strings
            
    s = ' '.join(new_string)                  # to rejoin the tokenized words with space
    s = s.strip()                             # to remove spaces before and after the whole string
    exclude = set(string.punctuation)         # remove punctuation
    s = ''.join(ch for ch in s if ch not in exclude)          #to join words who are not in exclude
    
    return s
```

```python
df['review_stop'] = df['reviews_emoji'].apply(transform_text)
```

```python
import re
def tokenize(data):
    split = re.split("\W+", data)         # to split the text into tokens
    return split

df['reviews_split'] = df['review_stop'].apply(lambda x: tokenize(x.lower))
```
