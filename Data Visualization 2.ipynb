```python
# 1. Use the dataset 'titanic'. Plot a box plot for distribution of age with respect to each gender along with the information
# about whether they survived or not. (Column names : 'sex' and 'age')
# 2. Write observations on the inference from the above statistics. Use voilin plot and others plots from seaborn package
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop('deck', inplace=True, axis=1)
```

```python
x=df
x1= df.dropna()
```

```python
x1.isnull().sum()
```

```python
sns.boxplot(data=x1, x='sex', y='age')
```

```python
sns.boxplot(data=x1, x='sex', y='age', hue='survived')
```

```python
sns.boxplot(data=x1, x='sex', y='age', hue='alive')
```

```python
sns.violinplot(data=x1, x='sex', y='age')
```

```python
sns.stripplot(data=x1, x='alive', y='age')
```

```python
sns.stripplot(data=x1, x='fare', y='sex')
```

```python
sns.swarmplot(data=x1, x='age', y='alive', hue='survived')
```
